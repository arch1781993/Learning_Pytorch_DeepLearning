```python
import torch
```

```python
my_tensor = torch.arange(10)
my_tensor
```

# Reshape and View

```python
my_tensor = my_tensor.reshape(2,5)
my_tensor
```

```python
# Reshaping if we dont know the number of items using -1. lot of times when data is huge we dont know how many items are there.
my_tensor2 = torch.arange(10)
my_tensor2 = my_tensor2.reshape(2,-1)
my_tensor2
```

```python
my_tensor2 = my_tensor2.reshape(-1,2)
my_tensor2
```

```python
# ReshaPING WITH view
my_tensor3 = torch.arange(10)
my_tensor4 = my_tensor3.view(2,5)
my_tensor4
```

```python
# CHeck out the differences between reshape and view
```

```python
# With reshape and view, the subsidiary tensors will update automatically if you update the base tensor
my_tensor5 = torch.arange(10)
my_tensor6 = my_tensor5.reshape(2,5)
my_tensor6
```

```python
# lets update my_tensor5
my_tensor5[1] = 3212
my_tensor5
```

```python
my_tensor6
```

# Slices

```python
my_tensor7 = torch.arange(10)
my_tensor7
```

```python
# Grabing an specific item
my_tensor7[7]            # Note: it is not a number, it is a tensor, think of it as a scalar tensor
```

```python
my_tensor8 = my_tensor7.reshape(5,2)
my_tensor8
```

```python
# Grabing a slice
my_tensor8[:,1]       # this returns a 1d kind of a list
```

```python
# what if we want to keep the column strucutre
my_tensor8[:,1:]
```
