# Pooling layers in CNN

image.png

In a Convolutional Neural Network (CNN), pooling is a downsampling operation that reduces the spatial dimensions (width and height) of a feature map. It typically follows a convolutional layer to simplify the data while retaining the most important information.  
Think of it as a way of "summarizing" the features found by the filters, making the network more efficient and robust.  
How Pooling Works:  
Pooling uses a sliding window (often $2 \times 2$) that moves across the input feature map with a certain stride. (In the above figure Max pool is used with a stride of 2)   
For each position of the window, it performs a mathematical operation to condense the values into a single number.

Why do we use Pooling?  
Dimensionality Reduction: By shrinking the map, it reduces the number of parameters and computational cost for the next layers.  
Translation Invariance: It helps the model recognize a feature (like an eye or a wheel) regardless of its exact pixel location. If the object shifts slightly, the "max" value in that region remains the same.  
Prevents Overfitting: By providing an abstract, lower-resolution version of the features, the model is less likely to "memorize" specific noise in the training data.

Common Types of Pooling:  
Max Pooling:	Selects the maximum value from the window.	The most popular type; it focuses on the most prominent features like edges.  
Average Pooling:	Calculates the average value of all pixels in the window.	Smooths the image; often used in the final layers of a network.  
Global Pooling	Reduces an entire feature map to a single value (max or average).	Used to replace fully connected layers at the end of a model to reduce parameters.
