# Tensor math operations
- Add, Subtract,Multiply, Divide, Remainder, Exponent
- Shorthand and Longhand
- Reassignment

```python
import torch
import numpy as np
```

```python
a = torch.tensor([1,2,3,4])
b = torch.tensor([5,6,7,8])
```

```python
# Addition (Shorthand)
a+b
```

```python
# Addition (Longhand)
torch.add(a,b)                    # shorthand, longhand both are same.
```

```python
# Subtractions (Shorthand)
b-a
```

```python
# Subtraction (Longhand)
torch.sub(b,a)
```

```python
# Multiplication (Shorthand)
a*b
```

```python
# Multiplication (Longhand)
torch.mul(a,b)
```

```python
# Division (Shorthand)
b/a
```

```python
# Division (Longhand)
torch.div(b,a)
```

```python
# Remainder (Shorthand)
b%a
```

```python
# Remainder (Longhand)
torch.remainder(b,a)
```

```python
# Exponent/Power (Longhand)
torch.pow(a,b)
```

```python
# For all mathematical operations (Another way to write Longhand)
a.add(b)
b.sub(a)
b.mul(a)
b.div(a)
b.remainder(a)
a.pow(b)
```

## Reassignment
- add an ( _ ) after the operation symbol (for longhand)

```python
# Shorthand
# a = a+b
```

```python
# Longhand
a.add_(b)
```
